# Linear Program Example
 ***
## Scenario
My GPA is the directly affected by how long I spend studying and working out; it goes up 0.1 for each hour I study, and 0.005 for each hour I work out.  I have to sleep for at least eight hours every day and, while my academic adviosr insists I maintain a minimum two-to-one ratio of studying to working out, my couch demands that I work out at least 1.5 hours per day.  Obviously, I can't spend a negative amount of time doing anything.
 
 ## Decision Variables
 $x_1$ = Number of hours per day spent studying<br>
 $x_2$ = Number of hours per day spent working out<br>


## Parameters
 $gpa_1$ = 0.01 (GPA boost for each hour spent studying)<br>
 $gpa_2$ = 0.005 (GPA boost for each hour spent working out)<br>
 ## Objective Function
 Max $Z = 0.01x_1 + 0.005x_2$<br>
 ## Subject To
 
 |   |   |
 |---:|:---|
 |$x_1+x_2$|$\le 16$<br>|
 |$-x_1+2x_2$|$\le 0$<br>|
 |$-x_2+1.5$|$\le 0$<br>|
 |$x_i$ |$\ge 0 \forall i$<br>| 

```python
# This is a code block that

import numpy as np
import matplotlib.pyplot as plt

#set size of graph
plt.figure(figsize=(10,10))

# Set x1 as "X" axis from 0 to 20 with values every .01
x1 = np.linspace(0, 20, 2000)
# Set x2 as "y" axis from 0 to 20 with values every .01
x2 = np.linspace(0, 20, 2000)

## Creating a different equation for each of the constraints and solving each for x2
# x1 + x2 <= 16
con1 = 16 - x1
# -x1 + 2x2 <= 0
con2 = x1/2
# -x2 + 1.5 <=0
con3 = 1.5

# Make plot
plt.plot(x1, con1, label=r'$x_1+x_2\leq16$', c='g') #plot constraint
plt.plot(x1, con2, label=r'$-x_1+2x_2\leq0$', c='b') #plot constraint
plt.axhline(con3, color='r', linestyle='-',label=r'$x_2\geq1.5$',) #plot a horizontal line


plt.xlim((0, 20)) #set x-axis bounds
plt.ylim((0, 20)) #set y-axis bounds
plt.xlabel(r'$x_1$') #set x-axis label
plt.ylabel(r'$x_2$') #set y-axis label


plt.legend(loc=3) #move plot legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=1, borderaxespad=0.) #format plot legend
```
