```python
import numpy as np
import matplotlib.pyplot as plt

#set size of graph
plt.figure(figsize=(10,10))

# x1 >= 0
x1 = np.linspace(0, 20, 2000)
# x2 >= 0
x2 = np.linspace(0, 20, 2000)
# z >= 0; used for objective function contours
Z = np.linspace(0, 20, 2000)

## Creating a different equation for each of the constraints and solving each for x2
# x2 <= 10
x2a = 10

# 2x1 + 5x2 <= 60
x2b = -2/5*x1 + 12

# x1 + x2 <= 18
x2c = -x1 + 18

# 3x1 + x2 <= 44
x2d = -3*x1 + 44

#define objective function
X1,X2 = np.meshgrid(x1,x2) #define 2D array of values to graph objective function
Z = 2*X1 + 1*X2 #actual objective function

# Make plot
plt.axhline(x2a, color='r', linestyle='-',label=r'$x_2\leq10$',) #plot a horizontal line
plt.plot(x1, x2b, label=r'$2x_1+5x_2\leq60$', c='g') #plot constraint
plt.plot(x1, x2c, label=r'$x_1+x_2\leq18$', c='b') #plot constraint
plt.plot(x1, x2d, label=r'$3x_1+x_2\leq44$', c='c') #plot constraint

plt.xlim((0, 20)) #set x-axis bounds
plt.ylim((0, 20)) #set y-axis bounds
plt.xlabel(r'$x_1$') #set x-axis label
plt.ylabel(r'$x_2$') #set y-axis label

#plot objective function contours
CS = plt.contour(x1,x2,Z, colors='k',linestyles='dashed') #plot objective function contours
plt.clabel(CS, inline=1, fontsize=10, fmt='%1.0f') #format contour labels
plt.title("Objective Function: "r'$Z=2x_1+x_2$') #set chart title

# Fill feasible region
x2e = np.minimum(np.minimum(x2a, x2b), np.minimum(x2c,x2d)) #define a function that finds minimum of constraints
plt.fill_between(x1, x2e, 0, where=x1>=0, color='grey', alpha=0.5) #fill in area between minimum of constraints and x-axis
plt.legend(loc=3) #move plot legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=1, borderaxespad=0.) #format plot legend

#plot points for simplex path
X1=[0,44/3,13,0,5,10] #list of x-axis values for points of interest
X2=[0,0,5,10,10,8] #list of y-axis values for points of interest
labels = ['','','(OPTIMAL)','','',''] #list of labels for pointe of interest
plt.plot(X1,X2,'ko',markersize=10) #plot points of interest
bbox_props = dict(boxstyle="round", fc="w", ec="0.5", alpha=0.9) #define label format for points of interest
for i in range(0,len(X1)): #graph labels for points of interest
    plt.annotate("("+str("{:.2f}".format(X1[i]))+","+str("{:.2f}".format(X2[i]))+"): Z="+
                 str("{:.2f}".format(2*X1[i]+1*X2[i]))+" "+labels[i], (X1[i]+.2, X2[i]+.2),bbox=bbox_props)
```
